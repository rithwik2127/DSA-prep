```python
def swapcase(text):
    result = ""
    for character in text:
        if "a" <= character <= "z":
            result += chr(ord(character) - 32) # Convert to uppercase
        elif "A" <= character <= "Z":
            result += chr(ord(character) + 32) # Convert to lowercase
        else:
            result += character
    return result
swapcase("vinay")
```

```python
def my_strip(s):
  chars= "*#@"
  start = 0
  end = len(s)-1

  while start <= end and s[start] in chars:
    start += 1
  while start <= end and s[end] in chars:
    end -= 1
  return s[start:end+1]
text = input("Enter a string:")
result = my_strip(text)
print(result)
```
